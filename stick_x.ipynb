```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
volts = np.array([0.03, 2.481, 4.095875])
soll = np.array([-1, 0, 1])
```

```python
plt.scatter(volts, soll)

plt.xlabel('Voltage [V]')
plt.ylabel('Soll')

plt.show()
```

```python
def parabel_fun(v, param_1, param_2):
    return param_1 * np.power(v, param_2) - 1
```

```python
best_params, _ = curve_fit(parabel_fun, volts, soll)
best_params
```

```python
x = np.linspace(0, 4.1, 100)
y = parabel_fun(x, best_params[0], best_params[1])

plt.scatter(volts, soll, c='r', marker='x')
plt.plot(x, y)

plt.show()
```
